````python
SUFFIXES = r"(limited|ltd|inc|llc|plc|corp(?:oration)?|company|co|group)"

def build_aliases(name: str, text: str):
    aliases = set()
    nm = (name or "").strip()
    if not nm:
      return aliases

    aliases.add(nm)

    # variant without corporate suffixes (for alias building only)
    nm_no_suf = re.sub(rf"\b{SUFFIXES}\b\.?", "", nm, flags=re.IGNORECASE).strip()
    if nm_no_suf and nm_no_suf.lower() != nm.lower():
        aliases.add(nm_no_suf)

    # acronym from first letters (e.g., Tata Consultancy Services -> TCS)
    words = re.findall(r"[A-Za-z0-9]+", nm_no_suf or nm)
    if len(words) >= 2:
        ac = "".join(w[0] for w in words).upper()
        if len(ac) >= 2:
            aliases.add(ac)

    # pull uppercase abbreviations/tickers from parentheses in the original text
    for par in re.findall(r"\(([^)]{1,40})\)", text or ""):
        for tok in re.findall(r"[A-Z]{2,6}", par):
            aliases.add(tok)

    return sorted(aliases, key=len, reverse=True)  # longer first

def clean_description(row):
    text = str(row.get("about", "") or "")
    name = str(row.get("name", "") or "")

    # 1) Remove URLs early
    text = re.sub(r"http\S+|www\.\S+", " ", text, flags=re.IGNORECASE)

    # 2) Build aliases BEFORE dropping parentheses so we can capture (TCS), (NASDAQ: CTSH), etc.
    aliases = build_aliases(name, text)

    # 3) Drop parenthetical blobs like (TCS), (NASDAQ: CTSH)
    text = re.sub(r"\([^)]*\)", " ", text)

    # 4) Remove exchange/ticker phrases that may appear outside parentheses
    text = re.sub(r"\b(NASDAQ|NYSE|BSE|NSE)\b[:\s]*[A-Z0-9.\-]{1,8}", " ", text, flags=re.IGNORECASE)

    # 5) Remove company name & aliases anywhere (with optional possessive â€™s/'s)
    if aliases:
        pattern = r"(?<!\w)(?:" + "|".join(map(re.escape, aliases)) + r")(?:'s|â€™s)?(?!\w)"
        text = re.sub(pattern, " ", text, flags=re.IGNORECASE)

    # 6) Whitespace/punctuation tidy
    text = re.sub(r"\s+", " ", text)
    text = re.sub(r"^\W+|\W+$", "", text)
    return text.strip()
````

````python
# Apply cleaning
df_cleaned["clean_about"] = df_cleaned.apply(clean_description, axis=1)
````

````python
from unsloth import FastLanguageModel
import torch

model_name = "unsloth/Phi-3-mini-4k-instruct-bnb-4bit"

max_seq_leght = 2048
dtype = None

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = model_name,
    max_seq_length = max_seq_leght,
    dtype = dtype,
    load_in_4bit = True,
)
````

````python
from datasets import Dataset

def format_prompt(example):
    return (
        f"### Input: Business Description: {example['input']}\n"
        f"### Output: {json.dumps({'company_name': example['output']})}<|endoftext|>"
    )

# Load your file
with open("companies_clean_1000.json", "r") as f:
    file = json.load(f)


formatted_data = [format_prompt(item) for item in file]
dataset = Dataset.from_dict({"text": formatted_data})

print(dataset[0]["text"])  # show first prompt
````

````python
model = FastLanguageModel.get_peft_model(
    model,
    r = 16,
    target_modules = [
        "q_proj", "k_proj", "v_proj", "o_proj",
        "gate_proj", "up_proj", "down_proj",
    ],
    lora_alpha = 32,
    lora_dropout = 0,
    bias = "none",
    use_gradient_checkpointing = "unsloth",
    random_state = 3407,
    use_rslora = False,
    loftq_config = None,
)
````

````python
from trl import SFTTrainer
from transformers import TrainingArguments


trainer = SFTTrainer(
    model=model,
    tokenizer=tokenizer,
    train_dataset=dataset,
    dataset_text_field="text",
    max_seq_length=2048,  # Lowered
    dataset_num_proc=2,
    args=TrainingArguments(
        per_device_train_batch_size=1,   # Lowered
        gradient_accumulation_steps=4,
        warmup_steps=10,
        num_train_epochs=3,
        learning_rate=2e-4,
        fp16=not torch.cuda.is_bf16_supported(),
        bf16=torch.cuda.is_bf16_supported(),
        logging_steps=25,
        optim="adamw_torch",             # Safer optimizer
        weight_decay=0.01,
        lr_scheduler_type="linear",
        seed=3407,
        output_dir="outputs",
        save_strategy="epoch",
        save_total_limit=2,
        dataloader_pin_memory=False,
        report_to="none",
    ),
)
````

````python
# Train the model
trainer_stats = trainer.train()
````

## Checking the Name is Avaliable Or Not

````python
product_description = input("Enter your business description: ")
````

````python
def check_domain(domain):
    if "." not in domain:
        domain += ".com"

    try:
        w = whois.whois(domain)

        if not w.domain_name:  # No info â†’ available
            return True
        return False
    except Exception as e:
        msg = str(e).lower()
        if "no match" in msg or "not found" in msg or "available" in msg:
            return True
        return False
````

````python
iteration = 1
while True:
    # Build prompt
    messages = [
        {"role": "user", "content": f"### Input: Business Description: {product_description}\n### Output:"}
    ]

    # Tokenize
    inputs = tokenizer.apply_chat_template(
        messages,
        tokenize=True,
        add_generation_prompt=True,
        return_tensors="pt",
    ).to("cuda")

    # Generate
    outputs = model.generate(
        input_ids=inputs,
        max_new_tokens=128,
        use_cache=True,
        temperature=0.7,
        do_sample=True,
        top_p=0.9,
    )

    # Decode
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)

    # Extract name
    if "### Output:" in response:
        candidate_name = response.split("### Output:")[-1].strip()
    else:
        candidate_name = response.strip()

    # Show iteration info
    print(f"Iteration {iteration}: Generated â†’ {candidate_name}")

    # Check domain
    if check_domain(candidate_name):
        product_name = candidate_name
        print(f"\nâœ… Found available domain: {product_name}\n")
        break
    else:
        print("âŒ Already registered. Trying again...\n")

    iteration += 1
````
